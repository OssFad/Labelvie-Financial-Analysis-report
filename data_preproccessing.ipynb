```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Financial Statement documents:


## Income Statement:


```python
income_statement= pd.read_excel(r'Finanacial Statement(LBV).xlsx',sheet_name='Income Statement')
income_statement=income_statement.iloc[2:,1:-1].reset_index(drop=True)

#Rename columns:
income_statement.columns=income_statement.columns.str.replace('FY','',regex=False)
income_statement.head()
```

## Balance sheet:


```python
balance_sheet=pd.read_excel(r'Finanacial Statement(LBV).xlsx',sheet_name='Balance Sheet')
balance_sheet=balance_sheet.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns
balance_sheet.columns=balance_sheet.columns.str.replace('FY','',regex=False)
balance_sheet.head()
```

## Cash Flow:


```python
cash_flow=pd.read_excel(r'Finanacial Statement(LBV).xlsx',sheet_name='Cash Flow')
cash_flow=cash_flow.iloc[2:,1:-1].reset_index(drop=True)

# Rename columns:
cash_flow.columns=cash_flow.columns.str.replace('FY','',regex=False)
cash_flow.head()
```

## Ratios:


```python
ratios=pd.read_excel(r'Finanacial Statement(LBV).xlsx',sheet_name='Ratios')
ratios=ratios.iloc[2:,1:-1].reset_index(drop=True)
# Drop Current column:
ratios=ratios.drop(columns='Current')

# Rename columns:
ratios.columns=ratios.columns.str.replace('FY','',regex=False)
ratios.head()
```

# Transpose data:


### Balance sheet Transpose :


```python
# Transpose data:
bs_trsp=balance_sheet.transpose().reset_index(drop=False)

# set first row as columns head:
bs_trsp.columns=bs_trsp.iloc[0]

# drop First row:
bs_trsp=bs_trsp.drop(0)
bs_trsp
```

```python
# convert first column to date type:
bs_trsp['Fiscal Year']=pd.to_datetime(bs_trsp['Fiscal Year'],errors='coerce')
#Change '2024-01-01' to  '2024-12-31':
bs_trsp['Fiscal Year']=bs_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
# re-convert to date type:
bs_trsp['Fiscal Year']=pd.to_datetime(bs_trsp['Fiscal Year'],errors='coerce')
bs_trsp
```

### Income statement data Transpose:


```python
# Income Statement:
is_trsp=income_statement.transpose().reset_index(drop=False)
# Set first row as column header:
is_trsp.columns=is_trsp.iloc[0]
# delete the first row:
is_trsp=is_trsp.drop(0)
is_trsp
```

```python
# convert first columns to date type:
is_trsp['Fiscal Year']=pd.to_datetime(is_trsp['Fiscal Year'],errors='coerce')
#Change '2024-01-01' to  '2024-12-31':
is_trsp['Fiscal Year']=is_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
# re-convert to data type:
is_trsp['Fiscal Year']=pd.to_datetime(is_trsp['Fiscal Year'],errors='coerce')
is_trsp
```

### cash flow Transpose:


```python
# Transpose cash flow data:
cf_trsp=cash_flow.transpose().reset_index(drop=False)
cf_trsp.columns=cf_trsp.iloc[0]
cf_trsp=cf_trsp.drop(0)
cf_trsp
```

```python
# Convert first colummn data type:
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'],errors='coerce')
cf_trsp['Fiscal Year']=cf_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
cf_trsp['Fiscal Year']=pd.to_datetime(cf_trsp['Fiscal Year'],errors='coerce')
cf_trsp
```

### Transpose Ratios data:


```python
rat_trsp=ratios.transpose().reset_index(drop=False)
rat_trsp.columns=rat_trsp.iloc[0]
rat_trsp=rat_trsp.drop(0)
rat_trsp
```

```python
# convert first column data type:
rat_trsp['Fiscal Year']=pd.to_datetime(rat_trsp['Fiscal Year'],errors='coerce')
rat_trsp['Fiscal Year']=rat_trsp['Fiscal Year'].dt.year.astype(str)+'-12-31'
rat_trsp['Fiscal Year']=pd.to_datetime(rat_trsp['Fiscal Year'],errors='coerce')
rat_trsp
```

# Preproccessing:


```python
# Create new variables:
##overview:
revenue=is_trsp['Revenue'].str.replace(',','').astype(float)
cogs=is_trsp['Cost of Revenue'].str.replace(',','').astype(float)
gross_profit=is_trsp['Gross Profit'].str.replace(',','').astype(float)
ebitda=is_trsp['EBITDA'].str.replace(',','').astype(float)
ebit=is_trsp['Operating Income'].str.replace(',','').astype(float)
ebt=is_trsp['Pretax Income'].str.replace(',','') .astype(float)
net_income=is_trsp['Net Income'].str.replace(',','').astype(float)
net_income_growth=is_trsp['Net Income Growth'].str.replace(',','').str.replace('%','').astype(float) / 100
earning_per_share=is_trsp['EPS (Basic)'].str.replace(',','').astype(float)
free_cash_flow=is_trsp['Free Cash Flow'].str.replace(',','').astype(float)
total_assets=bs_trsp['Total Assets'].str.replace(',','').astype(float)
total_liabilities=bs_trsp['Total Liabilities'].str.replace(',','').astype(float)
total_equity=bs_trsp['Shareholders\' Equity'].str.replace(',','').astype(float)
operating_cash_flow=cf_trsp['Operating Cash Flow'].str.replace(',','').astype(float)
investing_cash_flow=cf_trsp['Investing Cash Flow'].str.replace(',','').astype(float)
financing_cash_flow=cf_trsp['Financing Cash Flow'].str.replace(',','').astype(float)
net_cash_flow=cf_trsp['Net Cash Flow'].str.replace(',','').astype(float)
## activity ratios:
receivable=bs_trsp['Receivables'].str.replace(',','').astype(float)
begining_receivable= receivable.shift(-1,fill_value=np.mean(receivable))
average_receivable=(receivable+begining_receivable)/2
inventory=bs_trsp['Inventory'].str.replace(',','').astype(float)
begining_inventory=inventory.shift(-1,fill_value=np.mean(inventory))
average_inventory=(inventory+begining_inventory)/2
payable=bs_trsp['Accounts Payable'].str.replace(',','').astype(float)
begining_payable=payable.shift(-1,fill_value=np.mean(payable))
average_payable=(payable+begining_payable)/2
current_assets=bs_trsp['Total Current Assets'].str.replace(',','').astype(float)
current_liabilities=bs_trsp['Total Current Liabilities'].str.replace(',','').astype(float)
working_capital=current_assets-current_liabilities
begining_working_capital=working_capital.shift(-1,fill_value= np.mean(working_capital))
average_working_capital=(working_capital+begining_working_capital)/2
begining_equity=total_equity.shift(-1,fill_value=np.mean(total_equity))
average_equity=(total_equity+begining_equity)/2
## liquidity ratios :
cash_equivalent=bs_trsp['Cash & Equivalents'].str.replace(',','').astype(float)
## solvency ratios:
total_debt=bs_trsp['Total Debt'].str.replace(',','').astype(float)
begining_total_assets=total_assets.shift(-1,fill_value=np.mean(total_assets))
average_total_assets=(total_assets+begining_total_assets)/2
begining_total_liabilities=total_liabilities.shift(-1,fill_value=np.mean(total_liabilities))
average_total_liabilities=(total_liabilities+begining_total_liabilities)/2
interest_expenses=is_trsp['Interest Expense'].str.replace(',','').astype(float).abs()
## profitability:
return_on_equity_roe=rat_trsp['Return on Equity (ROE)'].str.replace(',','').str.replace('%','',regex=False).astype(float) /100
return_on_assets_roa=rat_trsp['Return on Assets (ROA)'].str.replace(',','').str.replace('%','',regex=False).astype(float)/100
return_on_invested_capital_roic=rat_trsp['Return on Capital (ROIC)'].str.replace(',','').str.replace('%','',regex=False).astype(float)/100
## valuation:
pe_ratio=rat_trsp['PE Ratio'].str.replace(',','').astype(float)
ps_ratio=rat_trsp['PS Ratio'].str.replace(',','').astype(float)
pb_ratio=rat_trsp['PB Ratio'].str.replace(',','').astype(float)
ev_ebitda=rat_trsp['EV/EBITDA Ratio'].str.replace(',','').astype(float)
ev_ebit=rat_trsp['EV/EBIT Ratio'].str.replace(',','').astype(float)
dividend_yield=rat_trsp['Dividend Yield'].str.replace(',','').str.replace('%','',regex=False).replace('-',np.nan).astype(float)/100

```

## Overview :


```python
overview=is_trsp[['Fiscal Year']].copy()
overview['Revenue']=revenue
overview['COGS']=cogs
overview['Gross Profit']=gross_profit
overview['EBITDA']=ebitda
overview['EBIT']=ebit
overview['EBT']=ebt
overview['Net Income']=net_income
overview['Net Income Growth']=net_income_growth
overview['EPS']=earning_per_share
overview['Net Cash Flow']=net_cash_flow
overview['Free Cash Flow']=free_cash_flow
overview['Total Assets']=total_assets
overview['Total Liabilities']=total_liabilities
overview['Total Equity']=total_equity
overview['Operating Cash Flow']=operating_cash_flow
overview['Investing Cash Flow']=investing_cash_flow
overview['Financing Cash Flow']=financing_cash_flow
overview
```

## Activity Ratios :


```python
activ_ratios=is_trsp[['Fiscal Year']].copy()
activ_ratios['Receivables']=receivable
activ_ratios['Inventory']=inventory
activ_ratios['Payables']=payable
activ_ratios['DSO']=(average_receivable/revenue)*365
activ_ratios['DIO']=(average_inventory/cogs)*365
activ_ratios['DPO']=(average_payable/cogs)*365
activ_ratios['CCC']= activ_ratios['DSO']+activ_ratios['DIO']-activ_ratios['DPO']
activ_ratios['Working Capital']=working_capital
activ_ratios['Working Capital Turnover']=revenue/average_working_capital
activ_ratios['Receivable Turnover']=revenue/average_receivable
activ_ratios['Iventory Turnover']=cogs/average_inventory
activ_ratios['Payable Turnover']=cogs/average_payable
activ_ratios['Equity Turnover']=revenue/average_equity
activ_ratios
```

## Liquidity Ratios :


```python
liqu_ratios=is_trsp[['Fiscal Year']].copy()
liqu_ratios['Current Ratio']=current_assets / current_liabilities
liqu_ratios['Quick Ratio']=(current_assets-inventory)/current_liabilities
liqu_ratios['Cash Ratio']=cash_equivalent/current_liabilities
liqu_ratios['Oprating Cash Flow Ratio']= operating_cash_flow/current_liabilities
liqu_ratios
```

## Solvency Ratios :


```python
solv_ratios=is_trsp[['Fiscal Year']].copy()
solv_ratios['Debt to Equity(D/E)']=total_debt/total_equity
solv_ratios['Debt to Capital(D/C)']=total_debt/(total_debt+total_equity)
solv_ratios['Debt to Assets']=total_debt/total_assets
solv_ratios['Financial Leverage']=average_total_assets/average_equity
solv_ratios['Interest coverage']=ebit/interest_expenses
solv_ratios
```

## Profitability Ratios :


```python
prof_ratios=is_trsp[['Fiscal Year']].copy()
prof_ratios['Gross Profit Margin']=gross_profit/revenue
prof_ratios['Operating Profit(EBIT) Margin']=ebit/revenue
prof_ratios['Pretax(EBT) Margin']=ebt/revenue
prof_ratios['Net Profit Margin']=net_income/revenue
prof_ratios['Return On Assets(ROA)']=return_on_assets_roa
prof_ratios['Return On Equity(ROE)']=return_on_equity_roe
prof_ratios['Return On Invested Capital(ROIC)']=return_on_invested_capital_roic
prof_ratios
```

## Valuation Ratios :


```python
val_ratios=is_trsp[['Fiscal Year']].copy()
val_ratios['Earnings per Share(EPS)']=earning_per_share
val_ratios['Price Earnings ratios (P/E)']=pe_ratio
val_ratios['Price to Sales(P/S)']=ps_ratio
val_ratios['Price to Book(P/B)']=pb_ratio
val_ratios['EV/EBITDA']=ev_ebitda
val_ratios['EV/EBIT']=ev_ebit
val_ratios['Dividend Yield']=dividend_yield
val_ratios
```
